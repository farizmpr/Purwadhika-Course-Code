Recommendation sytem

### 3 jenis utama recommendation system

### 1. democraphic filtering
- rekomendasi yang bersifat general
    - top 10 movies of the year
    - top 10 trending video in indonesia
    - top 50 global songs
- data rekomendasi berdasarkan demografik dan geografik user
- berdasarkan data awal dari user ketika melakukan registrasi / data awal yang diberikan hak ke apps atau web
- paling simple
- rekomendasi yang biasanya diberikan paling awal ketika user registrasi dan belum melakukan aktivitas


### 2. content-based filtering
- rekomendasi berdasarkan konten/produk yang diakses user
- rekomendasi bersifat spesifik (personalized)
    - misal marketplace: rekomendasi sesuai dengan barang yang pernah kita beli
    - youtube: sesuai dengan video yang pernah kita tonton
    - rekomendasi - iklan sesuai dengan barang/produk yang pernah kita search
    - netflix: rekomendasi berdasarkan genres, director, studio, dll
    
### 3. Collaborative filtering
- minimal ada 2 user yang melakukan aktivitas yang sama
- contohnya: narasi viewer also watch this, avneger viewer also watch this movie
- customer who viewed this product also viewed ....
- customer who bought this product also bought ...

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

## content based movie recommendation

```python
df=pd.read_csv('movie-dataset.csv')
df.head()
```

```python
df.isna().sum()
```

```python
# fitur disesuaikan dengan apa yang ingin direkomendasi berdasarkan apa, contohnya ini berdasarkan genres
fitur=['genres','keywords','cast','director']
```

```python
df.production_companies.unique()
```

```python
for i in fitur:
    df[i]=df[i].fillna('')
```

```python
df.isna().sum()
```

```python
# yang fitur dijadikan jadi satu
def combo(x):
    return x['genres']+" "+x['keywords']+" "+x['cast']+" "+x["director"]
```

```python
#combo feature
df['combo_features']=df.apply(combo,axis=1)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer# ngubah jadi vektor
```

```python
cv =CountVectorizer()
```

```python
movie_matrix=cv.fit_transform(df['combo_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_score=cosine_similarity(movie_matrix)#model recom itu cos score
```

```python
cos_score ==> Model
kalo recommendation yang di export itu cos score
```

### give recommendation

```python
fav_movie="John Carter"
```

```python
list(df[df['title']==fav_movie]['index'])[0]
```

```python
list(df[df['index']==2]['title'])[0]
```

```python
def index_by_movie(movie):
    return list(df[df['title']==movie]['index'])[0]
# ngeluarin judul
```

```python
def index_by_index(ind):
    return list(df[df['index']==ind]['title'])[0]
#ngeluarin index
```

```python
index_mov=index_by_movie(fav_movie)
```

```python
index_mov
```

```python
similiar_movie=list(enumerate(cos_score[index_mov]))
```

```python
similiar_movie
```

```python
similiar_movie[:6]
#mau mengurutkan berdasrakan cosinus similarity
```

```python
sorted_movies=sorted(similiar_movie,key=lambda x: x[1],reverse=True)[1:]
sorted_movies
```

```python
print(f"top 5 movie recommendation based on your movie :{fav_movie}")
count=0
for i in sorted_movies:
    print(index_by_index(i[0]))
    count+=1
    if count == 5:
        break
```
